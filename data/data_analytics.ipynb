```python
# data_analytics.csv

dfa = pd.read_csv('/content/data_analytics.csv', skiprows=9)
print(dfa.head())
```

```python
# data_survey.csv

dfs = pd.read_csv(
    '/content/data_survey.csv',
    sep=',',
    names=['timestamp', 'seen_pepe', 'platform', 'creator_control', 'explanation'],
    skiprows=1,
    encoding='latin-1'
)
```

```python
print("dfa shape:", dfa.shape)
print("dfs shape:", dfs.shape)
```

```python
# Analytics

fig, ax = plt.subplots(figsize=(8, 5))

plt.title('Figure 1: New Users by Channel Group (Feb 25 - Mar 24, 2026)')

channels = dfa['First user primary channel group (Default Channel Group)']
new_users = dfa['New users']

ax.bar(channels, new_users, color=['steelblue', 'seagreen'], edgecolor='black')

ax.set_ylabel('Number of New Users')
ax.set_xlabel('Channel Group')
ax.set_ylim(0, max(new_users) + 1)

plt.tight_layout()
plt.show()
```

```python
# Survey

fig, ax = plt.subplots(figsize=(8, 5))

plt.title('Figure 2: Where Do Students Usually See Memes?')

platforms = ['Instagram', 'TikTok', 'Reddit', 'Twitter']
counts = [4, 3, 1, 0]

ax.bar(platforms, counts, color=['coral', 'mediumpurple', 'orange', 'skyblue'], edgecolor='black')

ax.set_ylabel('Number of Respondents')
ax.set_xlabel('Platform')
ax.set_ylim(0, 5)

plt.tight_layout()
plt.show()
```

Introduction
This report presents findings from two datasets collected in March 2026: a Google Analytics report tracking website user acquisition, and a Google Forms survey examining student familiarity with Pepe the Frog memes and attitudes toward online content ownership.

Data Overview
The analytics dataset (dfa) has 2 rows and 5 columns, representing Direct and Organic Search traffic with metrics including new users, total revenue, transactions, and average 120-day value.

The survey dataset (dfs) has 4 rows and 5 columns, capturing each respondent's timestamp, whether they had seen Pepe the Frog memes, their platform, their view on creator control, and an optional explanation.

Figure 1: Website User Acquisition
Figure 1 shows new users by channel group from February 25 to March 24, 2026. The website had 4 new users total: 3 via Direct (75%) and 1 via Organic Search (25%). Revenue and transactions were zero, suggesting the site is in an early stage.

Figure 2: Where Students See Memes
Figure 2 shows platforms where respondents encounter memes. Instagram was most popular (4 respondents), followed by TikTok (3), Reddit (1), and Twitter (0).

Conclusion
Instagram and TikTok dominate as meme platforms. The majority of respondents believe creators should control their content online. The website is in early growth with only 4 new users recorded.
