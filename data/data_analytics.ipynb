```python
# data_survey.csv

# update the column names to match the names you have in your csv file for data_survey

# create a pandas dataframe
dfs = pd.read_csv(
    '/content/data_survey.csv',
    sep=',',
    names=['timestamp', 'seen_pepe', 'platform', 'creator_control', 'explanation'],
    skiprows=1,
    encoding='latin-1'
)





```

```python
print("dfa shape:", dfa.shape)
print("dfs shape:", dfs.shape)
```

```python
# Analytics
# make a blank canvas
fig, ax = plt.subplots(figsize=(8, 5))

# make a title starting with 'Figure 1:'
plt.title('Figure 1: New Users by Channel Group (Feb 25 - Mar 24, 2026)')

# plot
channels = dfa['First user primary channel group (Default Channel Group)']
new_users = dfa['New users']
colors = ['steelblue', 'seagreen']
ax.bar(channels, new_users, color=colors, edgecolor='black', width=0.5)

# labels
ax.set_ylabel('Number of New Users')
ax.set_xlabel('Channel Group')
ax.set_ylim(0, max(new_users) + 1)

# reveal everything
plt.tight_layout()
plt.show()


```

```python
# Survey


# make a blank canvas
fig, ax = plt.subplots(figsize=(8, 5))

# make a title starting with 'Figure 2:'
plt.title('Figure 2: Where Do Students Usually See Memes?')

# plot
platforms = ['Instagram', 'TikTok', 'Reddit', 'Twitter']
counts = [4, 3, 1, 0]
colors = ['coral', 'mediumpurple', 'orange', 'skyblue']
ax.bar(platforms, counts, color=colors, edgecolor='black')

# ylabel
ax.set_ylabel('Number of Respondents')

# xlabel
ax.set_xlabel('Platform')

# ylim
ax.set_ylim(0, 5)

# reveal everything
plt.tight_layout()
plt.show()



```
